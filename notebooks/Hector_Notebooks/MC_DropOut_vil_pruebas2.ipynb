### AQUI EMPIEZA LA VERDADERA COMPARACIÓN: BASE Vs MC DropOut

````python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# ---------------------------------------
# 1) Carga y preprocesamiento
df = pd.read_csv("train.csv")

num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']

# Mapear categóricas a enteros
df['Gender']    = df['Gender'].map({'Male':0,'Female':1})
df['Geography'] = df['Geography'].map({'France':0,'Spain':1,'Germany':2})

# Tipos y escalado
df[num_cols] = df[num_cols].astype(float)
df[cat_cols] = df[cat_cols].astype(int)
df['Exited']  = df['Exited'].astype(int)
scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Target y features numéricas/categóricas
y = df['Exited'].values.astype(np.float32)
X_num = df[num_cols]
X_cat = df[cat_cols]

# ---------------------------------------
# 2) Cálculo de Information Value (IV)
def calculate_iv(df, feature, target, bins=10):
    working = df[[feature, target]].copy()
    if feature in cat_cols:
        working['grp'] = working[feature]
    else:
        working['grp'] = pd.qcut(working[feature], bins, duplicates='drop')
    summary = working.groupby('grp').agg({target:['sum','count']})
    summary.columns = ['bad','total']
    summary['good'] = summary['total'] - summary['bad']
    summary['dist_bad']  = summary['bad']  / summary['bad'].sum()
    summary['dist_good'] = summary['good'] / summary['good'].sum()
    summary['woe'] = np.log(summary['dist_good'] / summary['dist_bad']).replace([np.inf,-np.inf],0)
    summary['iv']  = (summary['dist_good'] - summary['dist_bad']) * summary['woe']
    return summary['iv'].sum()

iv_values = {col: calculate_iv(df, col, 'Exited') for col in num_cols + cat_cols}
iv_series = pd.Series(iv_values).sort_values(ascending=False)
print("IV por variable:\n", iv_series)

# Selección numérica por IV
high_num = [c for c in num_cols if iv_series[c] >= 0.05]
low_num  = [c for c in num_cols if iv_series[c] <  0.05]

# ---------------------------------------
# 3) División train/test (estratificado)
X_num_tr, X_num_te, X_cat_tr, X_cat_te, y_tr, y_te = train_test_split(
    X_num, X_cat, y, test_size=0.15, stratify=y, random_state=42
)

# Convertir a numpy arrays
X_high_tr = X_num_tr[high_num].values.astype(np.float32)
X_low_tr  = X_num_tr[low_num].values.astype(np.float32)
X_high_te = X_num_te[high_num].values.astype(np.float32)
X_low_te  = X_num_te[low_num].values.astype(np.float32)
X_cat_tr_np = {c: X_cat_tr[c].values.astype(np.int32) for c in cat_cols}
X_cat_te_np = {c: X_cat_te[c].values.astype(np.int32) for c in cat_cols}

batch_size = 256

def make_ds(Xh, Xl, Xc, y):
    inputs = ((Xh, Xl), tuple(Xc[c] for c in cat_cols))
    return tf.data.Dataset.from_tensor_slices((inputs, y.astype(np.float32))) \
             .shuffle(len(y)).batch(batch_size).prefetch(tf.data.AUTOTUNE)

train_ds = make_ds(X_high_tr, X_low_tr, X_cat_tr_np, y_tr)
test_ds  = make_ds(X_high_te, X_low_te, X_cat_te_np, y_te)

# ---------------------------------------
# 4) Pesos empíricos
pos = float((y_tr==1).sum())
neg = float((y_tr==0).sum())
total = pos + neg
class_weight = {0:(1/neg)*(total/2), 1:(1/pos)*(total/2)}
print("Class weight empírico:", class_weight)

# ---------------------------------------
# 5) Focal Loss y EarlyStopping
def focal_loss(alpha, gamma=2.0):
    def loss(y_true, y_pred):
        y_t = tf.cast(y_true, tf.float32)
        p = tf.clip_by_value(y_pred,1e-7,1-1e-7)
        pt = y_t*p + (1-y_t)*(1-p)
        a  = y_t*alpha + (1-y_t)*(1-alpha)
        mod = tf.pow(1-pt, gamma)
        return tf.reduce_mean(-a * mod * tf.math.log(pt))
    return loss

alpha = 1.0 - (pos/total)
focal_loss_fn = focal_loss(alpha)
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)

# ---------------------------------------
# 6) Construcción de modelo con embeddings dinámicos
def build_model(use_mc_dropout=False):
    # Inputs
    inp_high = layers.Input(shape=(len(high_num),), name='num_high')
    inp_low  = layers.Input(shape=(len(low_num),), name='num_low')
    cat_inps = {c: layers.Input(shape=(), dtype='int32', name=c) for c in cat_cols}

    # Función para mapear IV a dimensión de embedding
    max_iv = iv_series.max()
    def emb_dim(iv):
        min_d, max_d = 1, 4
        return int(round(min_d + (max_d-min_d)*(iv/max_iv)))

    # Construir embeddings
    embs = []
    for c in cat_cols:
        d = emb_dim(iv_series[c])
        vn = df[c].nunique()
        e  = layers.Embedding(input_dim=vn, output_dim=d)(cat_inps[c])
        embs.append(layers.Flatten()(e))

    # Rama numérica alto IV
    x_high = inp_high
    for u in [256,128,64]:
        x_high = layers.Dense(u, activation='relu')(x_high)
        x_high = layers.Dropout(0.3)(x_high)

    # Rama numérica bajo IV
    x_low = inp_low
    for u in [32,16]:
        x_low = layers.Dense(u, activation='relu')(x_low)
        x_low = layers.Dropout(0.3)(x_low)

    # Concatenar todas las entradas
    x = layers.concatenate([x_high, x_low] + embs)
    for u in [32]:
        x = layers.Dense(u, activation='relu')(x)
        x = layers.Dropout(0.3)(x)
    out = layers.Dense(1, activation='sigmoid')(x)

    model = keras.Model(inputs=[inp_high, inp_low] + list(cat_inps.values()), outputs=out)
    return model

# ---------------------------------------
# 7) Entrenamiento y comparación
histories = {}
for name, mc in [('MLP', False), ('MC_Dropout', True)]:
    print(f"\n=== Entrenando {name} ===")
    model = build_model(use_mc_dropout=mc)
    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss=focal_loss_fn,
        metrics=['accuracy', keras.metrics.Recall(name='recall')]
    )
    history = model.fit(
        train_ds,
        validation_data=test_ds,
        epochs=50,
        class_weight=class_weight,
        callbacks=[early_stop],
        verbose=2
    )
    histories[name] = history
````

````python
# 8) Gráficas comparativas
def plot_comparison(histories, metric):
    plt.figure(figsize=(8,5))
    for name, h in histories.items():
        plt.plot(h.history[metric],      label=f"{name} train")
        plt.plot(h.history[f"val_{metric}"], '--', label=f"{name} val")
    plt.title(f"Comparación de modelos — {metric}")
    plt.xlabel("Época")
    plt.ylabel(metric)
    plt.legend(); plt.grid(True); plt.show()

plot_comparison(histories, 'loss')
plot_comparison(histories, 'accuracy')
plot_comparison(histories, 'recall')
````

````python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 1) Carga y preprocess
# ---------------------------------------
df = pd.read_csv("train.csv")
num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']
df['Gender']    = df['Gender'].map({'Male':0,'Female':1})
df['Geography'] = df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df['NumOfProducts'] = df['NumOfProducts'].astype(int) - 1   # 1→0, 2→1, 3→2, 4→3
df[num_cols] = df[num_cols].astype(float)
df[cat_cols] = df[cat_cols].astype(int)
df['Exited']  = df['Exited'].astype(int)
scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
y = df['Exited'].values.astype(np.float32)

# 2) IV y selección numérica
# ---------------------------------------
def calculate_iv(df, feature, target, bins=10):
    working = df[[feature,target]].copy()
    if feature in cat_cols:
        working['grp']=working[feature]
    else:
        working['grp']=pd.qcut(working[feature],bins,duplicates='drop')
    summary=working.groupby('grp').agg({target:['sum','count']})
    summary.columns=['bad','total']
    summary['good']=summary['total']-summary['bad']
    summary['dist_bad']=summary['bad']/summary['bad'].sum()
    summary['dist_good']=summary['good']/summary['good'].sum()
    summary['woe']=np.log(summary['dist_good']/summary['dist_bad']).replace([np.inf,-np.inf],0)
    summary['iv']=(summary['dist_good']-summary['dist_bad'])*summary['woe']
    return summary['iv'].sum()
iv_values={c:calculate_iv(df,c,'Exited') for c in num_cols+cat_cols}
iv_series=pd.Series(iv_values)
high_num=[c for c in num_cols if iv_series[c]>=0.05]
low_num=[c for c in num_cols if iv_series[c]<0.05]
print("High-IV numeric:",high_num)

# 3) Split Train/Test
# ---------------------------------------
X_high=df[high_num]; X_low=df[low_num]; X_cat_all=df[cat_cols]
Xh_tr,Xh_te,Xl_tr,Xl_te,Xcat_tr,Xcat_te,y_tr,y_te=train_test_split(
    X_high,X_low,X_cat_all,y,test_size=0.15,stratify=y,random_state=42)
Xh_tr,Xl_tr = Xh_tr.values.astype(np.float32), Xl_tr.values.astype(np.float32)
Xh_te,Xl_te = Xh_te.values.astype(np.float32), Xl_te.values.astype(np.float32)
Xcat_tr={c:Xcat_tr[c].values.astype(np.int32) for c in cat_cols}
Xcat_te={c:Xcat_te[c].values.astype(np.int32) for c in cat_cols}
batch_size=256
def make_ds(Xh,Xl,Xc,y):
    inputs=(Xh,Xl)+tuple(Xc[c] for c in cat_cols)
    return tf.data.Dataset.from_tensor_slices((inputs,y)).shuffle(len(y)).batch(batch_size).prefetch(tf.data.AUTOTUNE)
train_ds=make_ds(Xh_tr,Xl_tr,Xcat_tr,y_tr)
test_ds=make_ds(Xh_te,Xl_te,Xcat_te,y_te)

# 4) Class weight
# ---------------------------------------
pos,y_neg = y_tr.sum(),(len(y_tr)-y_tr.sum())
total=pos+y_neg
cw={0:(1/y_neg)*(total/2),1:(1/pos)*(total/2)}
print("Class weight:",cw)

# 5) Focal loss & EarlyStopping
# ---------------------------------------
def focal_loss(alpha,gamma=2.0):
    def loss(y_true,y_pred):
        y_t=tf.cast(y_true,tf.float32)
        p=tf.clip_by_value(y_pred,1e-7,1-1e-7)
        pt=y_t*p+(1-y_t)*(1-p)
        a=y_t*alpha+(1-y_t)*(1-alpha)
        mod=tf.pow(1-pt,gamma)
        return tf.reduce_mean(-a*mod*tf.math.log(pt))
    return loss
alpha=1-(pos/total)
focal_loss_fn=focal_loss(alpha)
early_stop=keras.callbacks.EarlyStopping(monitor='val_loss',patience=8,restore_best_weights=True)

# 6) Construcción modelo con MC Dropout
def build_mc():
    inp_h=layers.Input(shape=(len(high_num),)); inp_l=layers.Input(shape=(len(low_num),))
    cat_in=[layers.Input(shape=(),dtype='int32',name=c) for c in cat_cols]
    max_iv=iv_series.max(); emb_dim=lambda iv: max(1,int(round(1+3*(iv/max_iv))))
    embs=[layers.Flatten()(layers.Embedding(df[c].nunique(),emb_dim(iv_series[c]))(ci))
          for c,ci in zip(cat_cols,cat_in)]
    xh=inp_h; xl=inp_l
    for u in [256,128,64]: xh=layers.Dropout(0.3)(layers.Dense(u,'relu')(xh),training=True)
    for u in [32,16]:   xl=layers.Dropout(0.3)(layers.Dense(u,'relu')(xl),training=True)
    x=layers.concatenate([xh,xl]+embs)
    for u in [32]: x=layers.Dropout(0.3)(layers.Dense(u,'relu')(x),training=True)
    out=layers.Dense(1,'sigmoid')(x)
    return keras.Model(inputs=[inp_h,inp_l]+cat_in,outputs=out)

# 7) Entrenar
# ---------------------------------------
mc_model=build_mc()
mc_model.compile(optimizer='adam',loss=focal_loss_fn,metrics=['accuracy',keras.metrics.Recall(name='recall')])
mc_model.fit(train_ds,validation_data=test_ds,epochs=50,class_weight=cw,callbacks=[early_stop],verbose=2)

# 8) MC inference para incertidumbre
# ---------------------------------------
N=30; all_preds=[]
for _ in range(N): all_preds.append(mc_model.predict(test_ds,verbose=0).flatten())
all_preds=np.stack(all_preds,axis=0)  # shape (N,samples)
mean_preds=all_preds.mean(0); std_preds=all_preds.std(0)

# 9) DataFrame de incertidumbre
# ---------------------------------------
df_uncert=pd.DataFrame({'mean':mean_preds,'std':std_preds,'true':y_te})
thresh=df_uncert['std'].quantile(0.9)
df_uncert['reject']=df_uncert['std']>thresh
print(df_uncert.head())

# 10) Active Learning pool
# ---------------------------------------
top_k=100; al_pool=df_uncert.nlargest(top_k,'std')
print("Top-k uncertain indices:",al_pool.index.tolist())


````

````python
# =====================================================
# 0) Imports
# =====================================================
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# =====================================================
# 1) Carga y pre-proceso
# =====================================================
df = pd.read_csv("train.csv")

num_cols = ['CreditScore', 'Age', 'Tenure', 'Balance', 'EstimatedSalary']
cat_cols = ['Geography', 'Gender', 'NumOfProducts', 'HasCrCard', 'IsActiveMember']

df['Gender']    = df['Gender'].map({'Male': 0, 'Female': 1})
df['Geography'] = df['Geography'].map({'France': 0, 'Spain': 1, 'Germany': 2})
df['NumOfProducts'] = df['NumOfProducts'].astype(int) - 1

df[num_cols] = df[num_cols].astype(float)
df[cat_cols] = df[cat_cols].astype(int)
df['Exited'] = df['Exited'].astype(int)

scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

y = df['Exited'].values.astype(np.float32)

# =====================================================
# 2) Calcular IV para numéricas y categóricas
# =====================================================
def calculate_iv(data, feature, target, bins=10):
    work = data[[feature, target]].copy()
    if feature in cat_cols:
        work['grp'] = work[feature]
    else:
        work['grp'] = pd.qcut(work[feature], bins, duplicates='drop')
    s = work.groupby('grp').agg({target: ['sum', 'count']})
    s.columns = ['bad', 'total']
    s['good'] = s['total'] - s['bad']
    s['dist_bad']  = s['bad']  / s['bad'].sum()
    s['dist_good'] = s['good'] / s['good'].sum()
    s['woe'] = np.log(s['dist_good'] / s['dist_bad']).replace([np.inf, -np.inf], 0)
    s['iv']  = (s['dist_good'] - s['dist_bad']) * s['woe']
    return s['iv'].sum()

iv_vals = {c: calculate_iv(df, c, 'Exited') for c in num_cols + cat_cols}
iv_series = pd.Series(iv_vals)

high_num = [c for c in num_cols if iv_series[c] >= 0.05]
low_num  = [c for c in num_cols if iv_series[c]  < 0.05]

# =====================================================
# 3) Train/Test split → tf.data.Dataset
# =====================================================
X_high, X_low, X_cat_all = df[high_num], df[low_num], df[cat_cols]

Xh_tr, Xh_te, Xl_tr, Xl_te, Xcat_tr, Xcat_te, y_tr, y_te = train_test_split(
    X_high, X_low, X_cat_all, y, test_size=0.15, stratify=y, random_state=42)

Xh_tr, Xl_tr = Xh_tr.values.astype(np.float32), Xl_tr.values.astype(np.float32)
Xh_te, Xl_te = Xh_te.values.astype(np.float32), Xl_te.values.astype(np.float32)
Xcat_tr = {c: Xcat_tr[c].values.astype(np.int32) for c in cat_cols}
Xcat_te = {c: Xcat_te[c].values.astype(np.int32) for c in cat_cols}

batch_size = 256
def make_ds(Xh, Xl, Xc, y):
    inputs = (Xh, Xl) + tuple(Xc[c] for c in cat_cols)
    return (tf.data.Dataset.from_tensor_slices((inputs, y))
            .shuffle(len(y))
            .batch(batch_size)
            .prefetch(tf.data.AUTOTUNE))

train_ds = make_ds(Xh_tr, Xl_tr, Xcat_tr, y_tr)
test_ds  = make_ds(Xh_te, Xl_te, Xcat_te, y_te)

# =====================================================
# 4) Class weights para desbalanceo
# =====================================================
pos, neg = y_tr.sum(), len(y_tr) - y_tr.sum()
total = pos + neg
cw = {0: (1 / neg) * (total / 2), 1: (1 / pos) * (total / 2)}

# =====================================================
# 5) Focal Loss y EarlyStopping
# =====================================================
def focal_loss(alpha, gamma=2.0):
    def loss(y_true, y_pred):
        y_t = tf.cast(y_true, tf.float32)
        p   = tf.clip_by_value(y_pred, 1e-7, 1 - 1e-7)
        pt  = y_t * p + (1 - y_t) * (1 - p)
        a   = y_t * alpha + (1 - y_t) * (1 - alpha)
        mod = tf.pow(1 - pt, gamma)
        return tf.reduce_mean(-a * mod * tf.math.log(pt))
    return loss

alpha = 1 - (pos / total)
focal_loss_fn = focal_loss(alpha)

early_stop = keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=8, restore_best_weights=True, verbose=1
)

# =====================================================
# 6) Modelo MC-Dropout
# =====================================================
def build_mc():
    inp_h = layers.Input(shape=(len(high_num),))
    inp_l = layers.Input(shape=(len(low_num),))
    cat_inputs = [layers.Input(shape=(), dtype='int32', name=c) for c in cat_cols]

    max_iv = iv_series.max()
    emb_dim = lambda iv: max(1, int(round(1 + 3 * (iv / max_iv))))
    embs = [layers.Flatten()(layers.Embedding(df[c].nunique(), emb_dim(iv_series[c]))(ci))
            for c, ci in zip(cat_cols, cat_inputs)]

    xh = inp_h
    xl = inp_l
    for u in [256, 128, 64]:
        xh = layers.Dropout(0.3)(layers.Dense(u, 'relu')(xh), training=True)
    for u in [32, 16]:
        xl = layers.Dropout(0.3)(layers.Dense(u, 'relu')(xl), training=True)

    x = layers.concatenate([xh, xl] + embs)
    x = layers.Dropout(0.3)(layers.Dense(32, 'relu')(x), training=True)
    out = layers.Dense(1, 'sigmoid')(x)
    return keras.Model(inputs=[inp_h, inp_l] + cat_inputs, outputs=out)

mc_model = build_mc()
mc_model.compile(
    optimizer='adam',
    loss=focal_loss_fn,
    metrics=[
        metrics.BinaryAccuracy(name='accuracy'),
        metrics.Recall(name='recall'),
        metrics.AUC(name='auc', curve='ROC')
    ]
)

# =====================================================
# 7) Entrenamiento
# =====================================================
history = mc_model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=50,
    class_weight=cw,
    callbacks=[early_stop],
    verbose=2
).history

# =====================================================
# 8) Gráficas: loss, accuracy, recall, AUC
# =====================================================
def plot_history(h, metrics_list=("loss", "accuracy", "recall", "auc")):
    for m in metrics_list:
        plt.figure(figsize=(6, 4))
        plt.plot(h[m],            label="train")
        plt.plot(h["val_" + m], "--", label="val")
        plt.title(f"{m} — MC_Dropout")
        plt.xlabel("Época")
        plt.ylabel(m)
        plt.legend()
        plt.tight_layout()
        plt.show()

plot_history(history)

# =====================================================
# 9) Mejor AUC de validación
# =====================================================
best_ep = int(np.argmax(history["val_auc"]))
print(f"\nMejor val_AUC = {history['val_auc'][best_ep]:.4f}  (época {best_ep})")

# =====================================================
# 10) Inferencia MC para incertidumbre (opcional)
# =====================================================
N = 30
all_preds = []
for _ in range(N):
    all_preds.append(mc_model.predict(test_ds, verbose=0).flatten())

all_preds = np.stack(all_preds, axis=0)  # (N, muestras)
mean_preds = all_preds.mean(0)
std_preds  = all_preds.std(0)

df_uncert = pd.DataFrame({
    "mean": mean_preds,
    "std":  std_preds,
    "true": y_te
})
thresh = df_uncert["std"].quantile(0.9)
df_uncert["reject"] = df_uncert["std"] > thresh

print("\nMuestras con mayor incertidumbre (std > p90):")
print(df_uncert[df_uncert["reject"]].head())
````

````python
# =====================================================
# 0) Imports
# =====================================================
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# =====================================================
# 1) Carga y pre-proceso
# =====================================================
df = pd.read_csv("train.csv")

num_cols = ['CreditScore', 'Age', 'Tenure', 'Balance', 'EstimatedSalary']
cat_cols = ['Geography', 'Gender', 'NumOfProducts', 'HasCrCard', 'IsActiveMember']

df['Gender']    = df['Gender'].map({'Male': 0, 'Female': 1})
df['Geography'] = df['Geography'].map({'France': 0, 'Spain': 1, 'Germany': 2})
df['NumOfProducts'] = df['NumOfProducts'].astype(int) - 1

df[num_cols] = df[num_cols].astype(float)
df[cat_cols] = df[cat_cols].astype(int)
df['Exited'] = df['Exited'].astype(int)

scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
y = df['Exited'].values.astype(np.float32)

# =====================================================
# 2) Selección IV (igual que antes)
# =====================================================
def calculate_iv(data, feature, target, bins=10):
    work = data[[feature, target]].copy()
    if feature in cat_cols:
        work['grp'] = work[feature]
    else:
        work['grp'] = pd.qcut(work[feature], bins, duplicates='drop')
    s = work.groupby('grp').agg({target: ['sum', 'count']})
    s.columns = ['bad', 'total']
    s['good'] = s['total'] - s['bad']
    s['dist_bad']  = s['bad']  / s['bad'].sum()
    s['dist_good'] = s['good'] / s['good'].sum()
    s['woe'] = np.log(s['dist_good'] / s['dist_bad']).replace([np.inf, -np.inf], 0)
    s['iv']  = (s['dist_good'] - s['dist_bad']) * s['woe']
    return s['iv'].sum()

iv_vals   = {c: calculate_iv(df, c, 'Exited') for c in num_cols + cat_cols}
iv_series = pd.Series(iv_vals)

high_num = [c for c in num_cols if iv_series[c] >= 0.05]
low_num  = [c for c in num_cols if iv_series[c]  < 0.05]

# =====================================================
# 3) Train/Test split → tf.data.Dataset
# =====================================================
X_high, X_low, X_cat = df[high_num], df[low_num], df[cat_cols]
Xh_tr, Xh_te, Xl_tr, Xl_te, Xc_tr, Xc_te, y_tr, y_te = train_test_split(
    X_high, X_low, X_cat, y, test_size=0.15, stratify=y, random_state=42)

Xh_tr, Xl_tr = Xh_tr.values.astype(np.float32), Xl_tr.values.astype(np.float32)
Xh_te, Xl_te = Xh_te.values.astype(np.float32), Xl_te.values.astype(np.float32)
Xc_tr = {c: Xc_tr[c].values.astype(np.int32) for c in cat_cols}
Xc_te = {c: Xc_te[c].values.astype(np.int32) for c in cat_cols}

batch_size = 256
def make_ds(Xh, Xl, Xc, y):
    inputs = (Xh, Xl) + tuple(Xc[c] for c in cat_cols)
    return (tf.data.Dataset.from_tensor_slices((inputs, y))
            .shuffle(len(y)).batch(batch_size).prefetch(tf.data.AUTOTUNE))

train_ds = make_ds(Xh_tr, Xl_tr, Xc_tr, y_tr)
test_ds  = make_ds(Xh_te, Xl_te, Xc_te, y_te)

# =====================================================
# 4) Capa de temperatura
# =====================================================
class SigmoidTemperature(layers.Layer):
    """logits -> sigmoid(logits / T), con T entrenable (inicialmente 1)."""
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.log_T = self.add_weight(
            name='log_T', initializer='zeros', trainable=True)

    def call(self, logits):
        temperature = tf.exp(self.log_T)
        return tf.sigmoid(logits / temperature)

# =====================================================
# 5) Modelo MC-Dropout + temperatura
# =====================================================
def build_mc():
    inp_h = layers.Input(shape=(len(high_num),))
    inp_l = layers.Input(shape=(len(low_num),))
    cat_inputs = [layers.Input(shape=(), dtype='int32', name=c) for c in cat_cols]

    max_iv = iv_series.max()
    emb_dim = lambda iv: max(1, int(round(1 + 3 * (iv / max_iv))))
    embs = [layers.Flatten()(layers.Embedding(df[c].nunique(),
                                              emb_dim(iv_series[c]))(ci))
            for c, ci in zip(cat_cols, cat_inputs)]

    xh = inp_h
    xl = inp_l
    for u in [256, 128, 64]:
        xh = layers.Dropout(0.3)(layers.Dense(u, 'relu')(xh), training=True)
    for u in [32, 16]:
        xl = layers.Dropout(0.3)(layers.Dense(u, 'relu')(xl), training=True)

    x = layers.concatenate([xh, xl] + embs)
    x = layers.Dropout(0.3)(layers.Dense(32, 'relu')(x), training=True)

    logits = layers.Dense(1)(x)          # sin activación
    out    = SigmoidTemperature()(logits)
    return keras.Model(inputs=[inp_h, inp_l] + cat_inputs, outputs=out)

mc_model = build_mc()

# =====================================================
# 6) Compilación con label smoothing y Brier score
# =====================================================
bce = keras.losses.BinaryCrossentropy(label_smoothing=0.05)
brier = keras.metrics.MeanSquaredError(name='brier')

mc_model.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss=bce,
    metrics=[brier,
             metrics.BinaryAccuracy(name='accuracy'),
             metrics.Recall(name='recall'),
             metrics.AUC(name='auc', curve='ROC')]
)

early_stop = keras.callbacks.EarlyStopping(
    monitor='val_brier', patience=6, restore_best_weights=True, mode='min', verbose=1
)

# =====================================================
# 7) Entrenamiento
# =====================================================
history = mc_model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=50,
    callbacks=[early_stop],
    verbose=2
).history

# =====================================================
# 8) Gráficas de loss, brier y auc
# =====================================================
def plot_hist(h, metrics_to_show=("loss", "brier", "auc", "recall")):
    for m in metrics_to_show:
        plt.figure(figsize=(6,4))
        plt.plot(h[m],           label='train')
        plt.plot(h['val_'+m], '--', label='val')
        plt.title(f"{m} — MC-Dropout (calibración)")
        plt.xlabel("Época"); plt.ylabel(m); plt.legend(); plt.tight_layout(); plt.show()

plot_hist(history)

# =====================================================
# 9) Nueva curva de calibración y ECE
# =====================================================
from sklearn.calibration import calibration_curve
def ece(y_true, y_prob, k=10):
    bins = np.linspace(0,1,k+1)
    binids = np.digitize(y_prob, bins) - 1
    ece = 0
    for i in range(k):
        idx = binids == i
        if idx.sum() == 0: continue
        ece += abs(y_true[idx].mean() - y_prob[idx].mean()) * (idx.sum()/len(y_prob))
    return ece

probs = mc_model.predict(test_ds, verbose=0).flatten()
print("ECE:", ece(y_te, probs))

prob_true, prob_pred = calibration_curve(y_te, probs, n_bins=10, strategy="uniform")
plt.figure(figsize=(5,5))
plt.plot(prob_pred, prob_true, 'o-', label='MC-Dropout (calibrated)')
plt.plot([0,1],[0,1],'--',color='gray', label='Ideal')
plt.xlabel("Predicted probability"); plt.ylabel("Empirical probability")
plt.title("Calibration curve"); plt.legend(); plt.tight_layout(); plt.show()
````

````python
# =====================================================
# 0) Imports
# =====================================================
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
import tensorflow as tf
keras  = tf.keras;  layers = tf.keras.layers
from tensorflow.keras import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.isotonic import IsotonicRegression
from sklearn.metrics import (roc_auc_score, RocCurveDisplay,
                             PrecisionRecallDisplay, brier_score_loss,
                             confusion_matrix)
from sklearn.calibration import calibration_curve


plt.style.use("default")

# =====================================================
# 1) Carga y pre-proceso
# =====================================================
df = pd.read_csv("train.csv")

num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']

df['Gender']    = df['Gender'].map({'Male':0,'Female':1})
df['Geography'] = df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df['NumOfProducts'] = df['NumOfProducts'].astype(int)-1
df[num_cols] = df[num_cols].astype(float)
df[cat_cols] = df[cat_cols].astype(int)
df['Exited']  = df['Exited'].astype(int)

df[num_cols] = MinMaxScaler().fit_transform(df[num_cols])
y = df['Exited'].values.astype(np.float32)

# =====================================================
# 2) IV para seleccionar variables
# =====================================================
def calc_iv(data, feat, target='Exited', bins=10):
    tmp = data[[feat,target]].copy()
    tmp['grp'] = tmp[feat] if feat in cat_cols else pd.qcut(tmp[feat], bins, duplicates='drop')
    tab = tmp.groupby('grp').agg({target:['sum','count']}); tab.columns = ['bad','total']
    tab['good'] = tab['total']-tab['bad']
    tab['dist_bad']  = tab['bad']/tab['bad'].sum()
    tab['dist_good'] = tab['good']/tab['good'].sum()
    tab['woe'] = np.log(tab['dist_good']/tab['dist_bad']).replace([np.inf,-np.inf],0)
    tab['iv']  = (tab['dist_good']-tab['dist_bad'])*tab['woe']
    return tab['iv'].sum()

iv_series = pd.Series({c: calc_iv(df,c) for c in num_cols+cat_cols})
high_num  = [c for c in num_cols if iv_series[c] >= 0.05]
low_num   = [c for c in num_cols if iv_series[c]  < 0.05]

# =====================================================
# 3) Splits: TRAIN (60%) | CALIB (15%) | TEST (25 %)
# =====================================================
X_high, X_low, X_cat = df[high_num], df[low_num], df[cat_cols]
Xh_tmp, Xh_te, Xl_tmp, Xl_te, Xc_tmp, Xc_te, y_tmp, y_te = train_test_split(
    X_high, X_low, X_cat, y, test_size=0.25, stratify=y, random_state=42)
Xh_tr, Xh_cal, Xl_tr, Xl_cal, Xc_tr, Xc_cal, y_tr, y_cal = train_test_split(
    Xh_tmp, Xl_tmp, Xc_tmp, y_tmp, test_size=0.20, stratify=y_tmp, random_state=7)  # 0.20 de 0.75 ≈ 0.15

def to_np(Xh,Xl,Xc):    # (float32 / int32)
    return (Xh.values.astype(np.float32),
            Xl.values.astype(np.float32),
            {c:Xc[c].values.astype(np.int32) for c in cat_cols})

Xh_tr,Xl_tr,Xc_tr = to_np(Xh_tr,Xl_tr,Xc_tr)
Xh_cal,Xl_cal,Xc_cal = to_np(Xh_cal,Xl_cal,Xc_cal)
Xh_te,Xl_te,Xc_te = to_np(Xh_te,Xl_te,Xc_te)

def make_ds(Xh,Xl,Xc,y,shuffle):
    inputs = (Xh,Xl)+tuple(Xc[c] for c in cat_cols)
    ds = tf.data.Dataset.from_tensor_slices((inputs,y))
    if shuffle: ds = ds.shuffle(len(y))
    return ds.batch(256).prefetch(tf.data.AUTOTUNE)

train_ds = make_ds(Xh_tr,Xl_tr,Xc_tr,y_tr,shuffle=True)
calib_ds = make_ds(Xh_cal,Xl_cal,Xc_cal,y_cal,shuffle=False)
test_ds  = make_ds(Xh_te,Xl_te,Xc_te,y_te ,shuffle=False)

# =====================================================
# 4) Focal-Loss + EarlyStop
# =====================================================
pos,neg = y_tr.sum(), len(y_tr)-y_tr.sum(); total=pos+neg
alpha = 1-pos/total
def focal_loss(alpha,gamma=2.0):
    def loss(y_t,y_p):
        p = tf.clip_by_value(y_p,1e-7,1-1e-7)
        pt = y_t*p + (1-y_t)*(1-p)
        w  = y_t*alpha + (1-y_t)*(1-alpha)
        return tf.reduce_mean(-w*tf.pow(1-pt,gamma)*tf.math.log(pt))
    return loss
early = keras.callbacks.EarlyStopping(monitor='val_loss',patience=6,restore_best_weights=True,verbose=1)

# =====================================================
# 5) Modelo MC-Dropout
# =====================================================
def build_mc():
    in_h = layers.Input(shape=(len(high_num),))
    in_l = layers.Input(shape=(len(low_num),))
    cat_in = [layers.Input(shape=(),dtype='int32',name=c) for c in cat_cols]

    max_iv = iv_series.max(); embd = lambda iv:max(1,int(round(1+3*(iv/max_iv))))
    embs = [layers.Flatten()(layers.Embedding(df[c].nunique(),embd(iv_series[c]))(ci))
            for c,ci in zip(cat_cols,cat_in)]

    xh, xl = in_h, in_l
    for u in [256,128,64]: xh = layers.Dropout(0.3)(layers.Dense(u,'relu')(xh),training=True)
    for u in [32,16]:      xl = layers.Dropout(0.3)(layers.Dense(u,'relu')(xl),training=True)
    x = layers.concatenate([xh,xl]+embs)
    x = layers.Dropout(0.3)(layers.Dense(32,'relu')(x),training=True)
    out = layers.Dense(1,'sigmoid')(x)
    return keras.Model(inputs=[in_h,in_l]+cat_in,outputs=out)

mc_model = build_mc()
mc_model.compile(optimizer='adam',
                 loss=focal_loss(alpha),
                 metrics=[metrics.BinaryAccuracy(name='accuracy'),
                          metrics.Recall(name='recall'),
                          metrics.AUC(name='auc')])

# =====================================================
# 6) Entrenamiento
# =====================================================
hist = mc_model.fit(train_ds, validation_data=calib_ds,
                    epochs=50, callbacks=[early], verbose=2).history

# =====================================================
# 7) Funciones auxiliares
# =====================================================
def mc_pass_preds(model,dataset,n=30):
    passes=[]
    for _ in range(n):
        preds=[model(x,training=True).numpy().flatten() for (x,_) in dataset]
        passes.append(np.concatenate(preds))
    stack=np.stack(passes)            # (n, muestras)
    return stack.mean(0), stack.std(0)

def ece(y_true, y_prob, k=10):
    bins=np.linspace(0,1,k+1); idx=np.digitize(y_prob,bins)-1; e=0
    for i in range(k):
        m=idx==i
        if not m.any(): continue
        e += abs(y_prob[m].mean()-y_true[m].mean())*m.mean()
    return e

# =====================================================
# 8) MC-Dropout → media & std  (calib + test)
# =====================================================
mean_cal, std_cal = mc_pass_preds(mc_model, calib_ds)
mean_te , std_te  = mc_pass_preds(mc_model, test_ds )

# =====================================================
# 9) Calibrar la media con Isotonic
# =====================================================
iso = IsotonicRegression(out_of_bounds='clip').fit(mean_cal, y_cal)
mean_cal_iso = iso.transform(mean_cal)
mean_te_iso  = iso.transform(mean_te)

print("Brier antes :", brier_score_loss(y_te, mean_te))
print("Brier iso   :", brier_score_loss(y_te, mean_te_iso))
print("ECE antes   :", ece(y_te, mean_te))
print("ECE iso     :", ece(y_te, mean_te_iso))

# =====================================================
# 10) Thresholds:  σ (reject)  &  p (clasificación)
# =====================================================
rej_thr = np.quantile(std_cal, 0.90)      # rechazar 10 % más inciertos
prob_thr = 0.5                            # corte clásico (puedes optimizar con PR)

mask_acc = std_te <= rej_thr              # aceptados
y_acc    = y_te[mask_acc]
prob_acc = mean_te_iso[mask_acc]
pred_acc = (prob_acc >= prob_thr).astype(int)

# =====================================================
# 11) Graficar histórico
# =====================================================
for m in ("loss","accuracy","recall","auc"):
    plt.figure(figsize=(5,3))
    plt.plot(hist[m]); plt.plot(hist["val_"+m],'--')
    plt.title(m); plt.xlabel("epoch"); plt.ylabel(m); plt.tight_layout(); plt.show()

# =====================================================
# 12) Curva ROC & PR  (ACEPTADOS)
# =====================================================
RocCurveDisplay.from_predictions(y_acc, prob_acc)
plt.title("ROC (aceptados)"); plt.show()
PrecisionRecallDisplay.from_predictions(y_acc, prob_acc)
plt.title("PR (aceptados)"); plt.show()

# =====================================================
# 13) Calibration curve  (ACEPTADOS)
# =====================================================
pt, pp = calibration_curve(y_acc, prob_acc, n_bins=10, strategy="quantile")
plt.plot(pp, pt, 'o-'); plt.plot([0,1],[0,1],'--',c='gray')
plt.title("Calibration (aceptados)"); plt.xlabel("Predicted"); plt.ylabel("Empirical")
plt.tight_layout(); plt.show()

# =====================================================
# 14) Risk-Coverage  (accuracy)
# =====================================================
def risk_coverage(y, prob, std, metric="accuracy"):
    idx=np.argsort(std)            # std asc ⇒ menos incierto primero
    y,prob = y[idx], prob[idx]
    cover, quality = [], []
    for frac in np.linspace(0.1,1.0,10):
        m = int(len(y)*frac)
        y_sub, p_sub = y[:m], prob[:m]
        pred = (p_sub>=prob_thr).astype(int)
        if metric=="accuracy":
            q = (pred==y_sub).mean()
        else:
            from sklearn.metrics import f1_score
            q = f1_score(y_sub,pred)
        cover.append(frac); quality.append(q)
    return cover, quality

cov, acc = risk_coverage(y_te, mean_te_iso, std_te)
plt.plot(cov, acc, marker='o'); plt.xlabel("Cobertura"); plt.ylabel("Accuracy")
plt.title("Risk-Coverage (MC-Dropout)"); plt.tight_layout(); plt.show()

# =====================================================
# 15) Matriz de confusión (aceptados)
# =====================================================
cm = confusion_matrix(y_acc, pred_acc)
plt.figure(figsize=(4,3))
sns.heatmap(cm, annot=True, fmt='d', cmap='viridis', linewidths=1, linecolor='white')
plt.title("Confusion matrix (aceptados)"); plt.xlabel("Pred"); plt.ylabel("True")
plt.tight_layout(); plt.show()

print(f"\nCobertura   : {mask_acc.mean()*100:.1f}%")
print(f"Accuracy acc: {(pred_acc==y_acc).mean():.3f}")
print(f"AUC acc     : {roc_auc_score(y_acc, prob_acc):.3f}")
print(f"ECE acc     : {ece(y_acc, prob_acc):.4f}")
````

````python
# =====================================================
# 0) Imports
# =====================================================
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
import tensorflow as tf
keras, layers = tf.keras, tf.keras.layers
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.calibration import calibration_curve, IsotonicRegression
from sklearn.metrics import (roc_auc_score, precision_recall_curve,
                             RocCurveDisplay, PrecisionRecallDisplay,
                             brier_score_loss, confusion_matrix)

plt.style.use("default")

# =====================================================
# 1) Carga y pre-proceso
# =====================================================
df = pd.read_csv("train.csv")

num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']

df['Gender']    = df['Gender'].map({'Male':0,'Female':1})
df['Geography'] = df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df['NumOfProducts'] = df['NumOfProducts'].astype(int)-1
df[num_cols] = MinMaxScaler().fit_transform(df[num_cols].astype(float))
df[cat_cols] = df[cat_cols].astype(int)
df['Exited']  = df['Exited'].astype(int)
y = df['Exited'].values.astype(np.float32)

# =====================================================
# 2) Selección IV (opcional, mismo criterio que antes)
# =====================================================
def calc_iv(data, feat, target='Exited', bins=10):
    tmp = data[[feat,target]].copy()
    tmp['grp'] = tmp[feat] if feat in cat_cols else pd.qcut(tmp[feat], bins, duplicates='drop')
    tab = tmp.groupby('grp').agg({target:['sum','count']}); tab.columns=['bad','total']
    tab['good']=tab['total']-tab['bad']
    tab['dist_bad']=tab['bad']/tab['bad'].sum(); tab['dist_good']=tab['good']/tab['good'].sum()
    tab['woe']=np.log(tab['dist_good']/tab['dist_bad']).replace([np.inf,-np.inf],0)
    tab['iv']=(tab['dist_good']-tab['dist_bad'])*tab['woe']
    return tab['iv'].sum()

iv = pd.Series({c:calc_iv(df,c) for c in num_cols+cat_cols})
high_num = [c for c in num_cols if iv[c]>=0.05]
low_num  = [c for c in num_cols if iv[c]<0.05]

# =====================================================
# 3) Splits: 60 % train | 15 % calib | 25 % test
# =====================================================
Xh, Xl, Xc = df[high_num], df[low_num], df[cat_cols]
Xh_tmp,Xh_te,Xl_tmp,Xl_te,Xc_tmp,Xc_te,y_tmp,y_te = train_test_split(
    Xh,Xl,Xc,y,test_size=0.25,stratify=y,random_state=42)
Xh_tr,Xh_cal,Xl_tr,Xl_cal,Xc_tr,Xc_cal,y_tr,y_cal = train_test_split(
    Xh_tmp,Xl_tmp,Xc_tmp,y_tmp,test_size=0.2,stratify=y_tmp,random_state=7)

def to_np(Xh,Xl,Xc):
    return (Xh.values.astype(np.float32),
            Xl.values.astype(np.float32),
            {c:Xc[c].values.astype(np.int32) for c in cat_cols})

Xh_tr,Xl_tr,Xc_tr = to_np(Xh_tr,Xl_tr,Xc_tr)
Xh_cal,Xl_cal,Xc_cal = to_np(Xh_cal,Xl_cal,Xc_cal)
Xh_te,Xl_te,Xc_te = to_np(Xh_te,Xl_te,Xc_te)

def make_ds(Xh,Xl,Xc,y,shuffle):
    ds = tf.data.Dataset.from_tensor_slices(((Xh,Xl)+tuple(Xc[c] for c in cat_cols), y))
    if shuffle: ds = ds.shuffle(len(y))
    return ds.batch(256).prefetch(tf.data.AUTOTUNE)

train_ds = make_ds(Xh_tr,Xl_tr,Xc_tr,y_tr,shuffle=True)
calib_ds = make_ds(Xh_cal,Xl_cal,Xc_cal,y_cal,shuffle=False)
test_ds  = make_ds(Xh_te,Xl_te,Xc_te,y_te ,shuffle=False)

# =====================================================
# 4) Modelo MC-Dropout  (+ label smoothing)
# =====================================================
def build_mc():
    in_h,in_l = layers.Input((len(high_num),)), layers.Input((len(low_num),))
    cat_in = [layers.Input(shape=(),dtype='int32',name=c) for c in cat_cols]
    max_iv = iv.max(); emb=lambda iv_:max(1,int(round(1+3*(iv_/max_iv))))
    embs=[layers.Flatten()(layers.Embedding(df[c].nunique(),emb(iv[c]))(ci))
          for c,ci in zip(cat_cols,cat_in)]
    xh,xl=in_h,in_l
    for u in [256,128,64]: xh=layers.Dropout(0.3)(layers.Dense(u,'relu')(xh),training=True)
    for u in [32,16]:      xl=layers.Dropout(0.3)(layers.Dense(u,'relu')(xl),training=True)
    x=layers.concatenate([xh,xl]+embs)
    x=layers.Dropout(0.3)(layers.Dense(32,'relu')(x),training=True)
    out=layers.Dense(1,'sigmoid')(x)
    return keras.Model([in_h,in_l]+cat_in,out)

mc_model=build_mc()
mc_model.compile(optimizer='adam',
                 loss=keras.losses.BinaryCrossentropy(label_smoothing=0.05),
                 metrics=[keras.metrics.AUC(name='auc'),
                          keras.metrics.BinaryAccuracy(name='acc'),
                          keras.metrics.Recall(name='rec')])
early=keras.callbacks.EarlyStopping(monitor='val_auc',patience=6,
                                    restore_best_weights=True,mode='max',verbose=1)

# ⚡ Entrenamiento
hist=mc_model.fit(train_ds,validation_data=calib_ds,
                  epochs=50,callbacks=[early],verbose=2).history

# =====================================================
# 5) MC-Dropout — media & std  (calib + test)   ⚡
# =====================================================
def mc_pass(model,ds,n=30):
    outs=[np.concatenate([model(x,training=True).numpy().flatten()
                          for x,_ in ds]) for _ in range(n)]
    stack=np.stack(outs); return stack.mean(0),stack.std(0)
mean_cal,std_cal = mc_pass(mc_model,calib_ds)
mean_te ,std_te  = mc_pass(mc_model,test_ds )

# incertidumbre normalizada
unc_cal = std_cal/(mean_cal*(1-mean_cal)+1e-9)
unc_te  = std_te /(mean_te *(1-mean_te )+1e-9)

# =====================================================
# 6) Calibrar la media con Isotonic
# =====================================================
iso=IsotonicRegression(out_of_bounds='clip').fit(mean_cal,y_cal)
mean_te_iso = iso.transform(mean_te)

print("AUC test :", roc_auc_score(y_te, mean_te_iso))
print("Brier    :", brier_score_loss(y_te, mean_te_iso))

# =====================================================
# 7) Risk-Coverage (accuracy) usando incertidumbre
# =====================================================
def risk_cov(y,prob,unc,thr_list):
    acc=[]; cover=[]
    order=np.argsort(unc)          # asc  → menos incierto primero
    y,prob= y[order], prob[order]
    for thr in thr_list:
        m=int(len(y)*thr)
        pred=(prob[:m]>=0.5).astype(int)
        acc.append((pred==y[:m]).mean()); cover.append(thr)
    return cover,acc

thr_grid=np.linspace(0.1,1.0,10)
cov,acc=risk_cov(y_te,mean_te_iso,unc_te,thr_grid)
plt.plot(cov,acc,marker='o'); plt.xlabel("Cobertura"); plt.ylabel("Accuracy")
plt.title("Risk-Coverage (MC-Dropout, incertidumbre normalizada)")
plt.show()

# =====================================================
# 8) Elegir umbral de rechazo (p.ej. 90 % cobertura)
# =====================================================
rej_thr = np.percentile(unc_cal, 90)
mask_acc = unc_te <= rej_thr         # aceptados
y_acc    = y_te[mask_acc]
prob_acc = mean_te_iso[mask_acc]
pred_acc = (prob_acc>=0.5).astype(int)

print(f"Cobertura aceptados: {mask_acc.mean()*100:.1f}%")
print("AUC acc :", roc_auc_score(y_acc,prob_acc))
print("Brier acc :", brier_score_loss(y_acc,prob_acc))

# Confusión
cm=confusion_matrix(y_acc,pred_acc)
sns.heatmap(cm,annot=True,fmt='d',cmap='viridis');plt.title("Matriz (aceptados)")
plt.xlabel("Pred");plt.ylabel("True");plt.show()

# ROC y PR
RocCurveDisplay.from_predictions(y_acc,prob_acc);plt.title("ROC (aceptados)");plt.show()
PrecisionRecallDisplay.from_predictions(y_acc,prob_acc);plt.title("PR (aceptados)");plt.show()

# Curva de calibración (aceptados)
pt,pp=calibration_curve(y_acc,prob_acc,n_bins=10,strategy='quantile')
plt.plot(pp,pt,'o-'); plt.plot([0,1],[0,1],'--');plt.title("Calibración aceptados");plt.show()
````

````python
# =====================================================
# 0) Imports
# =====================================================
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
import tensorflow as tf
keras, layers = tf.keras, tf.keras.layers
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.calibration import IsotonicRegression, calibration_curve
from sklearn.metrics import (roc_auc_score, RocCurveDisplay,
                             PrecisionRecallDisplay, brier_score_loss,
                             confusion_matrix)

plt.style.use("default")
sns.set_context("notebook")

# =====================================================
# 1) Carga y pre-proceso
# =====================================================
df = pd.read_csv("train.csv")

num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']

df['Gender']    = df['Gender'].map({'Male':0,'Female':1})
df['Geography'] = df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df['NumOfProducts'] = df['NumOfProducts'].astype(int) - 1
df[num_cols] = MinMaxScaler().fit_transform(df[num_cols].astype(float))
df[cat_cols] = df[cat_cols].astype(int)
df['Exited']  = df['Exited'].astype(int)
y = df['Exited'].values.astype(np.float32)

# =====================================================
# 2) (Opcional) IV para seleccionar numéricas
# =====================================================
def calc_iv(data, feat, targ='Exited', bins=10):
    tmp = data[[feat,targ]].copy()
    tmp['grp'] = tmp[feat] if feat in cat_cols else pd.qcut(tmp[feat],bins,duplicates='drop')
    tab = tmp.groupby('grp').agg({targ:['sum','count']}); tab.columns=['bad','total']
    tab['good']=tab['total']-tab['bad']
    tab['dist_bad']=tab['bad']/tab['bad'].sum(); tab['dist_good']=tab['good']/tab['good'].sum()
    tab['woe']=np.log(tab['dist_good']/tab['dist_bad']).replace([np.inf,-np.inf],0)
    tab['iv']=(tab['dist_good']-tab['dist_bad'])*tab['woe']
    return tab['iv'].sum()

iv = pd.Series({c:calc_iv(df,c) for c in num_cols+cat_cols})
high_num = [c for c in num_cols if iv[c]>=0.05]
low_num  = [c for c in num_cols if iv[c]<0.05]

# =====================================================
# 3) Splits 60 % train | 15 % calib | 25 % test
# =====================================================
Xh, Xl, Xc = df[high_num], df[low_num], df[cat_cols]
Xh_tmp,Xh_te,Xl_tmp,Xl_te,Xc_tmp,Xc_te,y_tmp,y_te = train_test_split(
    Xh,Xl,Xc,y,test_size=0.25,stratify=y,random_state=42)
Xh_tr,Xh_cal,Xl_tr,Xl_cal,Xc_tr,Xc_cal,y_tr,y_cal = train_test_split(
    Xh_tmp,Xl_tmp,Xc_tmp,y_tmp,test_size=0.2,stratify=y_tmp,random_state=7)  # 0.2 de 0.75 ≈ 0.15

def to_np(Xh,Xl,Xc):
    return (Xh.values.astype(np.float32),
            Xl.values.astype(np.float32),
            {c:Xc[c].values.astype(np.int32) for c in cat_cols})
Xh_tr,Xl_tr,Xc_tr = to_np(Xh_tr,Xl_tr,Xc_tr)
Xh_cal,Xl_cal,Xc_cal = to_np(Xh_cal,Xl_cal,Xc_cal)
Xh_te,Xl_te,Xc_te = to_np(Xh_te,Xl_te,Xc_te)

def make_ds(Xh,Xl,Xc,y,shuffle=False):
    ds = tf.data.Dataset.from_tensor_slices(((Xh,Xl)+tuple(Xc[c] for c in cat_cols), y))
    if shuffle: ds = ds.shuffle(len(y))
    return ds.batch(256).prefetch(tf.data.AUTOTUNE)

train_ds = make_ds(Xh_tr,Xl_tr,Xc_tr,y_tr,shuffle=True)
calib_ds = make_ds(Xh_cal,Xl_cal,Xc_cal,y_cal)
test_ds  = make_ds(Xh_te,Xl_te,Xc_te,y_te)

# =====================================================
# 4) Modelo MC-Dropout con label_smoothing
# =====================================================
def build_mc():
    in_h, in_l = layers.Input((len(high_num),)), layers.Input((len(low_num),))
    cat_in = [layers.Input(shape=(), dtype='int32', name=c) for c in cat_cols]
    max_iv = iv.max(); embdim = lambda iv_: max(1,int(round(1+3*(iv_/max_iv))))
    embs = [layers.Flatten()(layers.Embedding(df[c].nunique(), embdim(iv[c]))(ci))
            for c,ci in zip(cat_cols, cat_in)]

    xh,xl = in_h,in_l
    for u in [256,128,64]:
        xh = layers.Dropout(0.3)(layers.Dense(u,'relu')(xh), training=True)
    for u in [32,16]:
        xl = layers.Dropout(0.3)(layers.Dense(u,'relu')(xl), training=True)

    x = layers.concatenate([xh,xl]+embs)
    x = layers.Dropout(0.3)(layers.Dense(32,'relu')(x), training=True)
    out = layers.Dense(1,'sigmoid')(x)
    return keras.Model([in_h,in_l]+cat_in, out)

mc_model = build_mc()
mc_model.compile(optimizer='adam',
                 loss=keras.losses.BinaryCrossentropy(label_smoothing=0.05),
                 metrics=[keras.metrics.AUC(name='auc'),
                          keras.metrics.BinaryAccuracy(name='acc'),
                          keras.metrics.Recall(name='rec')])

early = keras.callbacks.EarlyStopping(monitor='val_auc', patience=6,
                                      restore_best_weights=True, mode='max', verbose=1)

hist = mc_model.fit(train_ds, validation_data=calib_ds,
                    epochs=50, callbacks=[early], verbose=2).history

# =====================================================
# 5) Monte-Carlo passes → mean & entropy
# =====================================================
def mc_pass(model, ds, n=30):
    outs = [np.concatenate([model(x, training=True).numpy().flatten() for (x,_) in ds])
            for _ in range(n)]
    stack = np.stack(outs)
    mean, std = stack.mean(0), stack.std(0)
    return mean, std

mean_cal, std_cal = mc_pass(mc_model, calib_ds)
mean_te , std_te  = mc_pass(mc_model, test_ds)

eps = 1e-9
unc_cal = -(mean_cal*np.log(mean_cal+eps) + (1-mean_cal)*np.log(1-mean_cal+eps))
unc_te  = -(mean_te *np.log(mean_te +eps) + (1-mean_te )*np.log(1-mean_te +eps))

# =====================================================
# 6) Calibración isotónica
# =====================================================
iso = IsotonicRegression(out_of_bounds='clip').fit(mean_cal, y_cal)
mean_te_iso = iso.transform(mean_te)

print("AUC test :", roc_auc_score(y_te,  mean_te_iso))
print("Brier    :", brier_score_loss(y_te, mean_te_iso))

# =====================================================
# 7) Risk-Coverage con entropía
# =====================================================
order = np.argsort(unc_te)           # menos incertidumbre primero
cov, acc = [], []
for f in np.linspace(0.1,1.0,10):
    m = int(len(order)*f); idx = order[:m]
    cov.append(f)
    acc.append(((mean_te_iso[idx]>=0.5).astype(int)==y_te[idx]).mean())
plt.plot(cov, acc, marker='o')
plt.xlabel("Cobertura"); plt.ylabel("Accuracy")
plt.title("Risk-Coverage (entropía predictiva)")
plt.tight_layout(); plt.show()

# =====================================================
# 8) Umbral de rechazo (90 % cobertura)
# =====================================================
rej_thr = np.quantile(unc_cal, 0.90)
accept = unc_te <= rej_thr
y_acc  = y_te[accept]
p_acc  = mean_te_iso[accept]
pred   = (p_acc >= 0.5).astype(int)

print(f"\nCobertura: {accept.mean()*100:.1f}%")
print("AUC acc :",  roc_auc_score(y_acc, p_acc))
print("Brier acc:", brier_score_loss(y_acc, p_acc))

# Curvas y matriz
RocCurveDisplay.from_predictions(y_acc, p_acc)
plt.title("ROC (aceptados)"); plt.show()
PrecisionRecallDisplay.from_predictions(y_acc, p_acc)
plt.title("PR (aceptados)"); plt.show()
pt, pp = calibration_curve(y_acc, p_acc, n_bins=10, strategy="quantile")
plt.plot(pp,pt,'o-'); plt.plot([0,1],[0,1],'--'); plt.title("Calibración (aceptados)"); plt.show()
cm = confusion_matrix(y_acc, pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='viridis'); plt.title("Confusión (aceptados)"); plt.show()

# =====================================================
# 9) Función para inferencia con rechazo
# =====================================================
def predict_with_reject(x_inp, model, iso_model, unc_thr, n_pass=30, prob_thr=0.5):
    """Devuelve y_hat (-1 = reject), prob, uncertainty"""
    preds=[model(x_inp,training=True).numpy().flatten() for _ in range(n_pass)]
    preds=np.stack(preds)
    mean=preds.mean(0)
    unc = -(mean*np.log(mean+eps)+(1-mean)*np.log(1-mean+eps))
    prob=iso_model.transform(mean)
    out = np.full(mean.shape, -1)
    mask=unc<=unc_thr
    out[mask]=(prob[mask]>=prob_thr).astype(int)
    return out, prob, unc

# Ejemplo con el primer batch de test
sample_x, _ = next(iter(test_ds))
yhat, prob, unc = predict_with_reject(sample_x, mc_model, iso, rej_thr)
print("Prediciones (primer batch):", yhat[:15])
````

````python
# =====================================================
# 0) Imports
# =====================================================
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
import tensorflow as tf, warnings, os, math, random
keras, layers = tf.keras, tf.keras.layers
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.calibration import IsotonicRegression
from sklearn.metrics import roc_auc_score, brier_score_loss

plt.style.use("default");  sns.set_context("notebook")
warnings.filterwarnings("ignore")

# =====================================================
# 1) Datos y pre-proceso
# =====================================================
df = pd.read_csv("train.csv")
num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']

df['Gender']    = df['Gender'].map({'Male':0,'Female':1})
df['Geography'] = df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df['NumOfProducts'] = df['NumOfProducts'].astype(int)-1
df[num_cols] = MinMaxScaler().fit_transform(df[num_cols].astype(float))
df[cat_cols]  = df[cat_cols].astype(int)
df['Exited']  = df['Exited'].astype(int)
y = df['Exited'].values.astype(np.float32)

# ===== Information Value (rápido) para seleccionar numéricas
def calc_iv(data, feat, tgt='Exited', bins=10):
    x = data[[feat,tgt]].copy()
    x['grp'] = x[feat] if feat in cat_cols else pd.qcut(x[feat],bins,duplicates='drop')
    g = x.groupby('grp').agg({tgt:['sum','count']})
    g.columns=['bad','total']; g['good']=g['total']-g['bad']
    g['dist_bad']=g['bad']/g['bad'].sum(); g['dist_good']=g['good']/g['good'].sum()
    g['woe']=np.log(g['dist_good']/g['dist_bad']).replace([np.inf,-np.inf],0)
    g['iv']=(g['dist_good']-g['dist_bad'])*g['woe']
    return g['iv'].sum()
iv = pd.Series({c:calc_iv(df,c) for c in num_cols+cat_cols})
high_num = [c for c in num_cols if iv[c]>=0.05]
low_num  = [c for c in num_cols if iv[c]<0.05]

# =====================================================
# 2) Splits — 55 % train | 15 % calib | 15 % val | 15 % pool
# =====================================================
Xh, Xl, Xc = df[high_num], df[low_num], df[cat_cols]
Xh_tmp,Xh_pool,Xl_tmp,Xl_pool,Xc_tmp,Xc_pool,y_tmp,y_pool = train_test_split(
    Xh,Xl,Xc,y,test_size=0.15, stratify=y, random_state=42)       # pool 15 %
Xh_tr,Xh_val,Xl_tr,Xl_val,Xc_tr,Xc_val,y_tr,y_val = train_test_split(
    Xh_tmp,Xl_tmp,Xc_tmp,y_tmp,test_size=0.2143, stratify=y_tmp, random_state=7)  # val 15 % (= .15/.7)

def to_np(Xh,Xl,Xc):
    return (Xh.values.astype(np.float32),
            Xl.values.astype(np.float32),
            {c:Xc[c].values.astype(np.int32) for c in cat_cols})

Xh_tr,Xl_tr,Xc_tr = to_np(Xh_tr,Xl_tr,Xc_tr)
Xh_val,Xl_val,Xc_val = to_np(Xh_val,Xl_val,Xc_val)
Xh_pool,Xl_pool,Xc_pool = to_np(Xh_pool,Xl_pool,Xc_pool)

def make_ds(Xh,Xl,Xc,y,shuffle=False):
    ds = tf.data.Dataset.from_tensor_slices(((Xh,Xl)+tuple(Xc[c] for c in cat_cols), y))
    if shuffle: ds = ds.shuffle(len(y))
    return ds.batch(256).prefetch(tf.data.AUTOTUNE)

def rebuild_train_ds():          # helper usado más adelante
    global train_ds
    train_ds = make_ds(Xh_tr,Xl_tr,Xc_tr,y_tr,shuffle=True)

rebuild_train_ds()
val_ds   = make_ds(Xh_val,Xl_val,Xc_val,y_val)
pool_ds  = make_ds(Xh_pool,Xl_pool,Xc_pool,y_pool,shuffle=False)   # etiquetas solo para simulación

# =====================================================
# 3) Modelo MC-Dropout + funciones auxiliares
# =====================================================
def build_mc():
    in_h, in_l = layers.Input((len(high_num),)), layers.Input((len(low_num),))
    cat_in = [layers.Input(shape=(),dtype='int32',name=c) for c in cat_cols]
    max_iv = iv.max(); emb = lambda iv_: max(1,int(round(1+3*(iv_/max_iv))))
    embs = [layers.Flatten()(layers.Embedding(df[c].nunique(), emb(iv[c]))(ci))
            for c,ci in zip(cat_cols,cat_in)]
    xh,xl=in_h,in_l
    for u in [256,128,64]:
        xh = layers.Dropout(0.3)(layers.Dense(u,'relu')(xh), training=True)
    for u in [32,16]:
        xl = layers.Dropout(0.3)(layers.Dense(u,'relu')(xl), training=True)
    x = layers.concatenate([xh,xl]+embs)
    x = layers.Dropout(0.3)(layers.Dense(32,'relu')(x), training=True)
    out = layers.Dense(1,'sigmoid')(x)
    return keras.Model([in_h,in_l]+cat_in, out)

def compile_model(model):
    model.compile(optimizer='adam',
                  loss=keras.losses.BinaryCrossentropy(label_smoothing=0.05),
                  metrics=[keras.metrics.AUC(name='auc')])
    return model

def mc_mean_entropy(model, ds, n=30):
    outs=[np.concatenate([model(x,training=True).numpy().flatten() for (x,_) in ds])
          for _ in range(n)]
    preds=np.stack(outs)           # (n, muestras)
    mean = preds.mean(0)
    ent  = -(mean*np.log(mean+1e-9) + (1-mean)*np.log(1-mean+1e-9))
    return mean, ent

def iso_fit(mean_val, y_val):
    return IsotonicRegression(out_of_bounds='clip').fit(mean_val, y_val)

# =====================================================
# 4) Funciones Active-Learning
# =====================================================
def select_top_k(ent_pool, k):
    return np.argsort(ent_pool)[-k:]        # índices con mayor incertidumbre

def append_to_train(idx):
    """Añade los ejemplos indicados a train y actualiza el pool."""
    global Xh_tr,Xl_tr,Xc_tr,y_tr,Xh_pool,Xl_pool,Xc_pool,y_pool,pool_ds
    # 1) Añadir a train
    Xh_tr = np.vstack([Xh_tr, Xh_pool[idx]])
    Xl_tr = np.vstack([Xl_tr, Xl_pool[idx]])
    for c in cat_cols:
        Xc_tr[c] = np.concatenate([Xc_tr[c], Xc_pool[c][idx]])
    y_tr = np.concatenate([y_tr, y_pool[idx]])

    # 2) Crear máscara para mantener los NO seleccionados
    keep = np.ones(len(Xh_pool), dtype=bool)
    keep[idx] = False
    Xh_pool, Xl_pool, y_pool = Xh_pool[keep], Xl_pool[keep], y_pool[keep]
    for c in cat_cols:
        Xc_pool[c] = Xc_pool[c][keep]

    # 3) reconstruir datasets
    rebuild_train_ds()
    pool_ds = make_ds(Xh_pool, Xl_pool, Xc_pool, y_pool, shuffle=False)


def active_cycle(model, iso, k):
    # 1) MC sobre pool
    mean_pool, ent_pool = mc_mean_entropy(model, pool_ds)
    idx = select_top_k(ent_pool, k)
    # 2) (producción: etiquetar) – demo: usamos y_pool
    append_to_train(idx)
    # 3) Fine-tune
    compile_model(model)
    model.fit(train_ds, epochs=3, verbose=2,
              validation_data=val_ds)
    # 4) recalibrar isotónica
    mean_val,_ = mc_mean_entropy(model, val_ds)
    new_iso = iso_fit(mean_val, y_val)
    return model, new_iso

# =====================================================
# 5) Entrenamiento inicial
# =====================================================
mc_model = compile_model(build_mc())
early = keras.callbacks.EarlyStopping(monitor='val_auc', patience=6,
                                      mode='max', restore_best_weights=True, verbose=1)
mc_model.fit(train_ds, epochs=50, verbose=2, validation_data=val_ds, callbacks=[early])

mean_val,_ = mc_mean_entropy(mc_model, val_ds)
iso = iso_fit(mean_val, y_val)

print(f"Initial AUC val: {roc_auc_score(y_val, iso.transform(mean_val)):.4f}")

# =====================================================
# 6) 3 ciclos de Active-Learning
# =====================================================
CYCLES = 3
TOP_K  = 2000

for cycle in range(CYCLES):
    print(f"\n=== Active Cycle {cycle+1}/{CYCLES} ===")
    mc_model, iso = active_cycle(mc_model, iso, TOP_K)
    # Evalúa en validation
    mean_val,_ = mc_mean_entropy(mc_model, val_ds)
    auc_val = roc_auc_score(y_val, iso.transform(mean_val))
    print(f"AUC val después del ciclo {cycle+1}: {auc_val:.4f}")

# =====================================================
# 7) Métrica final en validation y pool restante
# =====================================================
mean_val,_ = mc_mean_entropy(mc_model, val_ds)
print("\nValidation  — AUC:", roc_auc_score(y_val, iso.transform(mean_val)),
      "  Brier:", brier_score_loss(y_val, iso.transform(mean_val)))

# pool restante como hold-out para ver mejora
if len(y_pool):
    pool_ds = make_ds(Xh_pool,Xl_pool,Xc_pool,y_pool)
    mean_pool,_ = mc_mean_entropy(mc_model, pool_ds)
    print("Pool restante — AUC:", roc_auc_score(y_pool, iso.transform(mean_pool)))
````

````python
# =====================================================
# 0) Imports
# =====================================================
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns, warnings
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.calibration import IsotonicRegression
from sklearn.metrics import roc_auc_score, brier_score_loss

warnings.filterwarnings("ignore")
keras, layers = tf.keras, tf.keras.layers
sns.set_context("notebook");  plt.style.use("default")

# =====================================================
# 1) Carga y pre-proceso
# =====================================================
df = pd.read_csv("train.csv")
num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']

df['Gender']    = df['Gender'].map({'Male':0,'Female':1})
df['Geography'] = df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df['NumOfProducts'] = df['NumOfProducts'].astype(int)-1
df[num_cols] = MinMaxScaler().fit_transform(df[num_cols].astype(float))
df[cat_cols] = df[cat_cols].astype(int)
df['Exited'] = df['Exited'].astype(int)
y = df['Exited'].values.astype(np.float32)

# ---------- IV para elegir numéricas ---------
def iv_of(feature):
    tmp = df[[feature,'Exited']].copy()
    tmp['grp'] = tmp[feature] if feature in cat_cols else pd.qcut(tmp[feature],10,duplicates='drop')
    g = tmp.groupby('grp').agg({'Exited':['sum','count']}); g.columns=['bad','total']
    g['good']=g['total']-g['bad']
    g['dist_bad']=g['bad']/g['bad'].sum(); g['dist_good']=g['good']/g['good'].sum()
    g['woe']=np.log(g['dist_good']/g['dist_bad']).replace([np.inf,-np.inf],0)
    g['iv']=(g['dist_good']-g['dist_bad'])*g['woe']
    return g['iv'].sum()
iv = pd.Series({c:iv_of(c) for c in num_cols+cat_cols})
high_num = [c for c in num_cols if iv[c]>=0.05]
low_num  = [c for c in num_cols if iv[c]<0.05]

# =====================================================
# 2) Splits 55 % train | 15 % calib | 15 % val | 15 % pool
# =====================================================
Xh,Xl,Xc = df[high_num], df[low_num], df[cat_cols]
Xh_tmp,Xh_pool,Xl_tmp,Xl_pool,Xc_tmp,Xc_pool,y_tmp,y_pool = train_test_split(
    Xh,Xl,Xc,y,test_size=0.15,stratify=y,random_state=0)

Xh_tr,Xh_val,Xl_tr,Xl_val,Xc_tr,Xc_val,y_tr,y_val = train_test_split(
    Xh_tmp,Xl_tmp,Xc_tmp,y_tmp,test_size=0.2143,stratify=y_tmp,random_state=1)      # 15 %
Xh_tr,Xh_cal,Xl_tr,Xl_cal,Xc_tr,Xc_cal,y_tr,y_cal = train_test_split(
    Xh_tr,Xl_tr,Xc_tr,y_tr,test_size=0.2143,stratify=y_tr,random_state=2)           # 15 %

def to_np(a,b,c):
    return (a.values.astype(np.float32),
            b.values.astype(np.float32),
            {k:c[k].values.astype(np.int32) for k in cat_cols})

Xh_tr,Xl_tr,Xc_tr = to_np(Xh_tr,Xl_tr,Xc_tr)
Xh_val,Xl_val,Xc_val = to_np(Xh_val,Xl_val,Xc_val)
Xh_cal,Xl_cal,Xc_cal = to_np(Xh_cal,Xl_cal,Xc_cal)
Xh_pool,Xl_pool,Xc_pool = to_np(Xh_pool,Xl_pool,Xc_pool)        # pool inicial

def make_ds(H,L,C,y,shuffle=False):
    ds = tf.data.Dataset.from_tensor_slices(((H,L)+tuple(C[k] for k in cat_cols), y))
    if shuffle: ds = ds.shuffle(len(y))
    return ds.batch(256).prefetch(tf.data.AUTOTUNE)

def rebuild_datasets():
    global train_ds, gold_ds, pool_ds
    train_ds = make_ds(Xh_tr,Xl_tr,Xc_tr,y_tr,shuffle=True)
    gold_ds  = make_ds(Xh_pool,Xl_pool,Xc_pool,y_pool)                    # etiquetas reales
    pool_ds  = make_ds(Xh_pool,Xl_pool,Xc_pool,np.zeros_like(y_pool))     # dummies

rebuild_datasets()
val_ds   = make_ds(Xh_val,Xl_val,Xc_val,y_val)
cal_ds   = make_ds(Xh_cal,Xl_cal,Xc_cal,y_cal)

# =====================================================
# 3) Modelo MC-Dropout
# =====================================================
def build_mc():
    h,l = layers.Input((len(high_num),)), layers.Input((len(low_num),))
    cat_in=[layers.Input(shape=(),dtype='int32') for _ in cat_cols]
    max_iv = iv.max(); emb = lambda v:max(1,int(round(1+3*(v/max_iv))))
    embs=[layers.Flatten()(layers.Embedding(df[c].nunique(),emb(iv[c]))(ci))
          for c,ci in zip(cat_cols,cat_in)]
    xh,xl=h,l
    for u in [256,128,64]: xh=layers.Dropout(0.3)(layers.Dense(u,'relu')(xh),training=True)
    for u in [32,16]: xl=layers.Dropout(0.3)(layers.Dense(u,'relu')(xl),training=True)
    x=layers.concatenate([xh,xl]+embs)
    x=layers.Dropout(0.3)(layers.Dense(32,'relu')(x),training=True)
    out=layers.Dense(1,'sigmoid')(x)
    return keras.Model([h,l]+cat_in,out)

def compile_(m):
    m.compile(optimizer='adam',
              loss=keras.losses.BinaryCrossentropy(label_smoothing=0.05),
              metrics=[keras.metrics.AUC(name='auc')])
    return m

def mc_mean_entropy(model,ds,n=30):
    outs=[np.concatenate([model(x,training=True).numpy().flatten() for x,_ in ds])
          for _ in range(n)]
    outs=np.stack(outs); mean=outs.mean(0)
    ent=-(mean*np.log(mean+1e-9)+(1-mean)*np.log(1-mean+1e-9))
    return mean, ent

# =====================================================
# 4) Entrenamiento inicial
# =====================================================
mc_model = compile_(build_mc())
mc_model.fit(train_ds,epochs=50,verbose=0,validation_data=cal_ds,
             callbacks=[keras.callbacks.EarlyStopping(monitor='val_auc',patience=6,
                                                     mode='max',restore_best_weights=True)])

mean_cal,_ = mc_mean_entropy(mc_model, cal_ds)
iso = IsotonicRegression(out_of_bounds='clip').fit(mean_cal, y_cal)

def auc_brier(ds, y_true):
    mean,_ = mc_mean_entropy(mc_model, ds)
    prob   = iso.transform(mean)
    return roc_auc_score(y_true, prob), brier_score_loss(y_true, prob)

def log_metrics(stage):
    val_auc,val_bri   = auc_brier(val_ds,  y_val)
    gold_auc,gold_bri = auc_brier(gold_ds, y_pool) if len(y_pool) else (np.nan,np.nan)
    print(f"{stage:>8} | AUC_val={val_auc:.4f}  Brier_val={val_bri:.4f} | "
          f"AUC_gold={gold_auc:.4f}")
    hist_val_auc.append(val_auc); hist_val_bri.append(val_bri)
    hist_gold_auc.append(gold_auc); hist_gold_bri.append(gold_bri)

hist_val_auc,hist_val_bri,hist_gold_auc,hist_gold_bri=[],[],[],[]
log_metrics("Start")

# =====================================================
# 5) Active Learning
# =====================================================
CYCLES, TOP_K = 3, 2000
for cyc in range(1,CYCLES+1):
    print(f"\n=== Cycle {cyc}/{CYCLES} ===")
    mean_pool, ent_pool = mc_mean_entropy(mc_model, pool_ds)
    sel_idx = np.argsort(ent_pool)[-TOP_K:]            # más inciertos
    # --- añadir a train ---
    global Xh_tr,Xl_tr,Xc_tr,y_tr,Xh_pool,Xl_pool,Xc_pool,y_pool
    Xh_tr = np.vstack([Xh_tr, Xh_pool[sel_idx]])
    Xl_tr = np.vstack([Xl_tr, Xl_pool[sel_idx]])
    for c in cat_cols:
        Xc_tr[c] = np.concatenate([Xc_tr[c], Xc_pool[c][sel_idx]])
    y_tr = np.concatenate([y_tr, y_pool[sel_idx]])
    keep = np.ones(len(Xh_pool), dtype=bool); keep[sel_idx] = False
    Xh_pool,Xl_pool,y_pool = Xh_pool[keep], Xl_pool[keep], y_pool[keep]
    for c in cat_cols:  Xc_pool[c] = Xc_pool[c][keep]
    rebuild_datasets()

    # --- fine-tune ---
    compile_(mc_model)
    mc_model.fit(train_ds, epochs=3, verbose=0, validation_data=cal_ds)

    # --- recalibrar ---
    mean_cal,_ = mc_mean_entropy(mc_model, cal_ds)
    iso = iso.fit(mean_cal, y_cal)

    log_metrics(f"Cycle{cyc}")

# =====================================================
# 6) Evolución de métricas
# =====================================================
plt.plot(hist_val_auc,'o-'); plt.title("AUC validation"); plt.show()
plt.plot(hist_val_bri,'o-'); plt.gca().invert_yaxis(); plt.title("Brier validation"); plt.show()
plt.plot(hist_gold_auc,'o-'); plt.title("AUC gold-test"); plt.show()
````

````python
# =====================================================
# 0) Imports
# =====================================================
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import (roc_auc_score, brier_score_loss,
                             confusion_matrix, precision_recall_curve)
from sklearn.calibration import calibration_curve
from sklearn.linear_model import LogisticRegression
from sklearn.isotonic import IsotonicRegression
import matplotlib.pyplot as plt
import seaborn as sns

# =====================================================
# 1) (Solo si aún no tienes los datos listos)
#     Carga, IV y Dataset — MISMA LÓGICA QUE TU CÓDIGO ORIGINAL
# =====================================================
df = pd.read_csv("train.csv")
num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']
df['Gender']    = df['Gender'].map({'Male':0,'Female':1})
df['Geography'] = df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df['NumOfProducts'] = df['NumOfProducts'].astype(int) - 1
df[num_cols] = df[num_cols].astype(float)
df[cat_cols] = df[cat_cols].astype(int)
df['Exited']  = df['Exited'].astype(int)
scaler = MinMaxScaler();  df[num_cols] = scaler.fit_transform(df[num_cols])
y = df['Exited'].values.astype(np.float32)

def calculate_iv(data, feature, target, bins=10):
    work = data[[feature, target]].copy()
    if feature in cat_cols: work['grp']=work[feature]
    else:                   work['grp']=pd.qcut(work[feature],bins,duplicates='drop')
    s = work.groupby('grp').agg({target:['sum','count']})
    s.columns=['bad','total'];   s['good']=s['total']-s['bad']
    s['dist_bad']=s['bad']/s['bad'].sum(); s['dist_good']=s['good']/s['good'].sum()
    s['woe']=np.log(s['dist_good']/s['dist_bad']).replace([np.inf,-np.inf],0)
    s['iv']=(s['dist_good']-s['dist_bad'])*s['woe'];  return s['iv'].sum()

iv_vals = {c:calculate_iv(df,c,'Exited') for c in num_cols+cat_cols}
iv_series=pd.Series(iv_vals)
high_num=[c for c in num_cols if iv_series[c]>=0.05]
low_num =[c for c in num_cols if iv_series[c]<0.05]

# =====================================================
# 2) Primer split: Train/Test (15 % test)
# =====================================================
Xh, Xl, Xc = df[high_num], df[low_num], df[cat_cols]
Xh_tr, Xh_te, Xl_tr, Xl_te, Xc_tr, Xc_te, y_tr, y_te = train_test_split(
    Xh, Xl, Xc, y, test_size=0.15, stratify=y, random_state=42)

# =====================================================
# 3) Segundo split: de Train separa “calib” (25 %)
#    ── se usará SOLO para ajustar isotónica / Platt
# =====================================================
Xh_tr, Xh_cal, Xl_tr, Xl_cal, Xc_tr, Xc_cal, y_tr, y_cal = train_test_split(
    Xh_tr, Xl_tr, Xc_tr, y_tr, test_size=0.25, stratify=y_tr, random_state=12)

# =====================================================
# 4) Convierte a numpy (float32 / int32) y arma tf.data
# =====================================================
def to_numpy_split(Xh,Xl,Xc):
    Xh = Xh.values.astype(np.float32)
    Xl = Xl.values.astype(np.float32)
    Xc = {c:Xc[c].values.astype(np.int32) for c in cat_cols}
    return Xh,Xl,Xc

Xh_tr,Xl_tr,Xc_tr = to_numpy_split(Xh_tr,Xl_tr,Xc_tr)
Xh_cal,Xl_cal,Xc_cal = to_numpy_split(Xh_cal,Xl_cal,Xc_cal)
Xh_te,Xl_te,Xc_te = to_numpy_split(Xh_te,Xl_te,Xc_te)

batch_size=256
def make_ds(Xh,Xl,Xc,y,shuffle=False):
    inputs=(Xh,Xl)+tuple(Xc[c] for c in cat_cols)
    ds=tf.data.Dataset.from_tensor_slices((inputs,y))
    if shuffle: ds=ds.shuffle(len(y))
    return ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)

train_ds = make_ds(Xh_tr,Xl_tr,Xc_tr,y_tr,shuffle=True)
calib_ds = make_ds(Xh_cal,Xl_cal,Xc_cal,y_cal,shuffle=False)
test_ds  = make_ds(Xh_te,Xl_te,Xc_te,y_te,shuffle=False)

# =====================================================
# 5) (Supón que tu modelo MC-Dropout original ya está entrenado)
#    Si NO lo tienes, aquí tienes una función mínima para recrearlo
#    (idéntica arquitectura previa con focal-loss y class_weight)
# =====================================================
def build_mc_original():
    inp_h=layers.Input(shape=(len(high_num),))
    inp_l=layers.Input(shape=(len(low_num),))
    cat_in=[layers.Input(shape=(),dtype='int32',name=c) for c in cat_cols]
    max_iv=iv_series.max(); embdim=lambda iv:max(1,int(round(1+3*(iv/max_iv))))
    embs=[layers.Flatten()(layers.Embedding(df[c].nunique(),embdim(iv_series[c]))(ci))
          for c,ci in zip(cat_cols,cat_in)]
    xh,xl=inp_h,inp_l
    for u in [256,128,64]: xh=layers.Dropout(0.3)(layers.Dense(u,'relu')(xh),training=True)
    for u in [32,16]:      xl=layers.Dropout(0.3)(layers.Dense(u,'relu')(xl),training=True)
    x=layers.concatenate([xh,xl]+embs)
    x=layers.Dropout(0.3)(layers.Dense(32,'relu')(x),training=True)
    out=layers.Dense(1,'sigmoid')(x)
    return keras.Model(inputs=[inp_h,inp_l]+cat_in,outputs=out)

# — ENTRENAR (solo si no lo tenías ya):
# cw = {0: ..., 1: ...}  # tus class_weight originales
# mc_model = build_mc_original()
# mc_model.compile(optimizer='adam', loss=focal_loss_fn,
#                  metrics=['accuracy', keras.metrics.Recall(name='recall'), keras.metrics.AUC(name='auc')])
# mc_model.fit(train_ds, validation_data=calib_ds, epochs=... , class_weight=cw)

# =====================================================
# 6) Probabilidades sin calibrar
# =====================================================
probs_cal = mc_model.predict(calib_ds, verbose=0).flatten()
probs_te  = mc_model.predict(test_ds , verbose=0).flatten()

# =====================================================
# 7) Isotonic Regression
# =====================================================
iso = IsotonicRegression(out_of_bounds='clip')
iso.fit(probs_cal, y_cal)
probs_iso = iso.transform(probs_te)

# =====================================================
# 8) Platt scaling (Logistic Regression)
# =====================================================
platt = LogisticRegression(solver='lbfgs')
platt.fit(probs_cal.reshape(-1,1), y_cal)
probs_platt = platt.predict_proba(probs_te.reshape(-1,1))[:,1]

# =====================================================
# 9) Funciones de métricas
# =====================================================
def ece(y_true, y_prob, k=10):
    bins = np.linspace(0,1,k+1);  idx=np.digitize(y_prob,bins)-1; e=0
    for i in range(k):
        mask = idx==i
        if not mask.any(): continue
        e += abs(y_prob[mask].mean()-y_true[mask].mean())*mask.mean()
    return e

def quick_stats(name, y_true, y_prob):
    auc = roc_auc_score(y_true,y_prob)
    bri = brier_score_loss(y_true,y_prob)
    print(f"{name:<12} | AUC {auc:.4f}  | Brier {bri:.4f} | ECE {ece(y_true,y_prob):.4f}")

# =====================================================
# 10) Comparación de métricas globales
# =====================================================
quick_stats("Original", y_te, probs_te)
quick_stats("Isotonic", y_te, probs_iso)
quick_stats("Platt",    y_te, probs_platt)

# =====================================================
# 11) Curvas de calibración
# =====================================================
plt.figure(figsize=(5,5))
for lab, pr in [("Original",probs_te), ("Isotonic",probs_iso), ("Platt",probs_platt)]:
    p_true, p_pred = calibration_curve(y_te, pr, n_bins=10, strategy="quantile")
    plt.plot(p_pred, p_true, marker='o', label=lab)
plt.plot([0,1],[0,1],'--',color='gray',label='Ideal')
plt.title("Calibration curves"); plt.xlabel("Predicted"); plt.ylabel("Empirical")
plt.legend(); plt.tight_layout(); plt.show()

# =====================================================
# 12) Matriz de confusión con umbral 0.5 (Original vs Isotonic)
#      – adapta “pred_thr” si prefieres otro corte.
# =====================================================
thr = 0.5
for lab, pr in [("Original",probs_te), ("Isotonic",probs_iso)]:
    y_pred = (pr >= thr).astype(int)
    cm = confusion_matrix(y_te,y_pred)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='viridis',
                linewidths=1, linecolor='white')
    plt.title(f"Confusion matrix — {lab}")
    plt.xlabel("Pred"); plt.ylabel("True"); plt.tight_layout(); plt.show()

# =====================================================
# 13) Umbral óptimo F1 tras calibrar (ejemplo con Isotonic)
# =====================================================
prec, rec, thr_pr = precision_recall_curve(y_te, probs_iso)
f1 = 2*prec*rec/(prec+rec+1e-12)
thr_f1 = thr_pr[np.argmax(f1)]
print(f"Isotonic — threshold F1 óptimo: {thr_f1:.3f}")
````

````python
import numpy as np
import tensorflow as tf

def mc_predictions(model, dataset, n_passes=30):
    """
    Devuelve la media de n_passes forward-passes con dropout activo.
    """
    all_passes = []
    for _ in range(n_passes):
        pass_preds = []
        for x_batch, _ in dataset:                 # ‼️  x_batch = tuple de inputs
            y_hat = model(x_batch, training=True)  # activa Dropout
            pass_preds.append(y_hat.numpy().flatten())
        all_passes.append(np.concatenate(pass_preds))
    return np.mean(all_passes, axis=0)   # media MC (shape = nº muestras)

# --- usa tu tf.data.Dataset de test ---
probs_mc = mc_predictions(mc_model, test_ds, n_passes=30)
````

````python
# ---------- CONFIGURACIÓN ----------
num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']
batch_size = 256
N_MC       = 30        # número de pasadas para dropout
cycles     = 3         # rondas de active learning
batch_AL   = 20        # muestras que se etiquetan por ronda

# ---------- CARGA ----------
df = pd.read_csv("train.csv")

df['Gender']      = df['Gender'].map({'Male':0,'Female':1})
df['Geography']   = df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df['NumOfProducts']= df['NumOfProducts'].astype(int) - 1   # 1→0 …

df[num_cols] = MinMaxScaler().fit_transform(df[num_cols])

# ---------- CALCULA IV ----------
def calculate_iv(df, feature, target, bins=10):
    working = df[[feature,target]].copy()
    if feature in cat_cols:
        working['grp'] = working[feature]
    else:
        working['grp'] = pd.qcut(working[feature], bins, duplicates='drop')
    s = working.groupby('grp').agg({target:['sum','count']})
    s.columns = ['bad','total']
    s['good'] = s['total'] - s['bad']
    s['dist_bad']  = s['bad']  / s['bad'].sum()
    s['dist_good'] = s['good'] / s['good'].sum()
    s['woe'] = np.log(s['dist_good'] / s['dist_bad']).replace([np.inf,-np.inf],0)
    s['iv']  = (s['dist_good'] - s['dist_bad']) * s['woe']
    return s['iv'].sum()

iv_values = {c: calculate_iv(df,c,'Exited') for c in num_cols+cat_cols}
iv_series = pd.Series(iv_values)
high_num  = [c for c in num_cols if iv_series[c] >= 0.05]
low_num   = [c for c in num_cols if iv_series[c] <  0.05]

# ---------- MATRIZ FINAL ----------
X_full = df[high_num + low_num + cat_cols].astype("float32").values
y_full = df['Exited'].astype("int32").values
````

````python
def build_mc():
    inp_h = tf.keras.Input(shape=(len(high_num),))
    inp_l = tf.keras.Input(shape=(len(low_num),))
    cat_in= [tf.keras.Input(shape=(), dtype='int32', name=c) for c in cat_cols]

    max_iv = iv_series.max()
    dim = lambda iv: max(1, int(round(1 + 3*(iv/max_iv))))

    embs = [tf.keras.layers.Flatten()(
                tf.keras.layers.Embedding(df[c].nunique(), dim(iv_series[c]))(ci))
            for c,ci in zip(cat_cols,cat_in)]

    xh = inp_h
    for u in [256,128,64]:
        xh = tf.keras.layers.Dense(u, activation='relu')(xh)
        xh = tf.keras.layers.Dropout(0.3)(xh, training=True)

    xl = inp_l
    for u in [32,16]:
        xl = tf.keras.layers.Dense(u, activation='relu')(xl)
        xl = tf.keras.layers.Dropout(0.3)(xl, training=True)

    x = tf.keras.layers.concatenate([xh, xl] + embs)
    x = tf.keras.layers.Dense(32, activation='relu')(x)
    x = tf.keras.layers.Dropout(0.3)(x, training=True)
    out = tf.keras.layers.Dense(1, activation='sigmoid')(x)

    model = tf.keras.Model(inputs=[inp_h, inp_l] + cat_in, outputs=out)
    model.compile(optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.Recall(name='recall')])
    return model
````

````python
class MCDropoutClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, epochs=20, batch_size=256, N_mc=30, verbose=0):
        self.epochs=epochs; self.batch_size=batch_size
        self.N_mc=N_mc; self.verbose=verbose

    # ---------- helpers ----------
    def _split(self, X):
        nH, nL, nC = len(high_num), len(low_num), len(cat_cols)
        Xh = X[:, :nH]
        Xl = X[:, nH:nH+nL]
        Xc = [X[:, nH+nL+i].astype("int32") for i in range(nC)]
        return (Xh, Xl, *Xc)

    # ---------- scikit-learn API ----------
    def fit(self, X, y):
        self.model_ = build_mc()
        self.model_.fit(self._split(X), y,
                        epochs=self.epochs,
                        batch_size=self.batch_size,
                        verbose=self.verbose)
        return self

    def predict_mc(self, X):
        preds = [self.model_.predict(self._split(X),
                                     batch_size=self.batch_size,
                                     verbose=0).flatten()
                 for _ in range(self.N_mc)]
        return np.stack(preds, axis=1)           # shape (n, N_mc)

    def predict_proba(self, X):
        p1 = self.predict_mc(X).mean(axis=1, keepdims=True)   # (n,1)
        p0 = 1.0 - p1                                         # (n,1)
        return np.concatenate([p0, p1], axis=1)               # (n,2)


    def predict(self, X):
        return (self.predict_proba(X)[:,1] >= 0.5).astype(int)
````

````python
from sklearn.base import BaseEstimator, ClassifierMixin

class MCDropoutClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, build_fn, epochs=50, batch_size=256,
                 class_weight=None, N_mc=30, verbose=0):
        self.build_fn      = build_fn
        self.epochs        = epochs
        self.batch_size    = batch_size
        self.class_weight  = class_weight
        self.N_mc          = N_mc
        self.verbose       = verbose

    # scikit-learn API ------------
    def fit(self, X, y):
        # X llega concatenado; separamos igual que antes
        Xh, Xl = X[:, :len(high_num)], X[:, len(high_num):len(high_num)+len(low_num)]
        Xc     = [X[:, -len(cat_cols)+i].astype("int32") for i in range(len(cat_cols))]
        self.model_ = self.build_fn()
        self.model_.compile(optimizer='adam', loss=focal_loss_fn, metrics=['accuracy'])
        self.model_.fit((Xh, Xl, *Xc), y,
                        epochs=self.epochs,
                        batch_size=self.batch_size,
                        class_weight=self.class_weight,
                        verbose=self.verbose)
        return self

    def predict_proba(self, X):
        Xh, Xl = X[:, :len(high_num)], X[:, len(high_num):len(low_num)+len(high_num)]
        Xc     = [X[:, -len(cat_cols)+i].astype("int32") for i in range(len(cat_cols))]
        preds  = []
        for _ in range(self.N_mc):
            preds.append(self.model_.predict((Xh, Xl, *Xc),
                                             batch_size=self.batch_size,
                                             verbose=0).flatten())
        return np.stack(preds, axis=1).mean(axis=1, keepdims=True)   # shape (n,1)

    # para compatibilidad con ActiveLearner
    def predict(self, X):
        p = self.predict_proba(X)[:, 0]
        return (p >= 0.5).astype(int)
````

````python
import numpy as np, pandas as pd, tensorflow as tf
from scikeras.wrappers import KerasClassifier
from modAL.models import ActiveLearner
from modAL.uncertainty import uncertainty_sampling
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# --- 0)  Load & preprocess ---------------------------------------------------
df = pd.read_csv("train.csv")

num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']

df['Gender']    = df['Gender'].map({'Male':0,'Female':1})
df['Geography'] = df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df[num_cols] = MinMaxScaler().fit_transform(df[num_cols])

X = df[num_cols + cat_cols].astype("float32").values
y = df['Exited'].astype("int32").values

# --- 1)  Split labelled vs. pool (your 100 uncertain indices) ---------------
uncertain_idx = np.array([3436, 19278, 22093, 13122, 15075, 5457, 23320, 15118,
                      16429, 6230, 3497, 7117, 14117, 8568, 23159, 22213,
                      830, 12319, 13595, 15225, 11632, 17297, 11904, 4768,
                      14356, 2169, 18861, 18285, 20326, 11117, 9389, 18865,
                      6450, 5350, 19224, 17946, 126, 18269, 16755, 16774,
                      7269, 4380, 6425, 89, 9446, 20517, 14607, 1759, 5989,
                      17301, 6997, 4501, 5481, 16714, 6226, 1066, 13510,
                      23326, 19399, 15880, 19151, 18554, 20411, 5585, 22330,
                      17529, 22756, 8361, 20452, 24153, 15429, 23643, 20164,
                      20979, 18680, 7252, 3378, 2816, 4454, 11477, 15090,
                      6022, 7748, 1235, 5123, 13982, 6362, 8505, 9995,
                      3493, 17444, 10349, 24458, 19813, 24550, 2482, 21078,
                      17668, 22666, 21978])
mask_pool     = np.zeros(len(df), bool)
mask_pool[uncertain_idx] = True

X_train, y_train = X[~mask_pool], y[~mask_pool]   # initially labelled
X_pool,  pool_id = X[mask_pool],  uncertain_idx   # to be queried

# --- 2)  Small Keras network wrapped for scikit-learn ------------------------
def build_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(64, activation="relu", input_shape=(X.shape[1],)),
        tf.keras.layers.Dense(32, activation="relu"),
        tf.keras.layers.Dense(1,  activation="sigmoid")
    ])
    model.compile(optimizer="adam",
                  loss="binary_crossentropy",
                  metrics=["accuracy"])
    return model

clf = KerasClassifier(model=build_model,
                      epochs=30,
                      batch_size=256,
                      verbose=0)

# --- 3)  Active-learning loop -----------------------------------------------
learner = ActiveLearner(
    estimator=clf,
    query_strategy=uncertainty_sampling,
    X_training=X_train, y_training=y_train
)

for cycle in range(10):                      # ask in batches of 10
    ask_idx, ask_X = learner.query(X_pool, n_instances=10)
    print(f"Cycle {cycle+1}: label samples with original row-ids ->",
          pool_id[ask_idx])

    # ⬇️ here you, the human, look at those rows and assign labels -------
    newly_labelled_y = y[pool_id[ask_idx]]

    # teach and remove from pool
    learner.teach(ask_X, newly_labelled_y)
    X_pool  = np.delete(X_pool, ask_idx, axis=0)
    pool_id = np.delete(pool_id, ask_idx)
````

### METEMOS EL CONJUNTO TEST PARA HACER PRUEBAS DE INCERTIDUMBRE.

````python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import MinMaxScaler

# 1) Carga y preprocesamiento de test.csv
df_test = pd.read_csv("test.csv")

num_cols = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
cat_cols = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']

# Mapear categóricas
df_test['Gender']    = df_test['Gender'].map({'Male':0,'Female':1})
df_test['Geography'] = df_test['Geography'].map({'France':0,'Spain':1,'Germany':2})

# Escalar continuas con el scaler entrenado
scaler = MinMaxScaler()
scaler.fit(pd.read_csv("train.csv")[num_cols])  # o carga tu scaler guardado
df_test[num_cols] = scaler.transform(df_test[num_cols])

# 2) Preparar inputs según high_num / low_num usados en entrenamiento
high_num = ['Age','Balance']            # ejemplo; usa tu propia lista
low_num  = [c for c in num_cols if c not in high_num]

Xh = df_test[high_num].values.astype(np.float32)
Xl = df_test[low_num].values.astype(np.float32)
Xc = {c: df_test[c].values.astype(np.int32) for c in cat_cols}

# 3) Crear Dataset sin etiquetas
batch_size = 256
def make_test_ds(Xh, Xl, Xc):
    inputs = (Xh, Xl) + tuple(Xc[c] for c in cat_cols)
    return tf.data.Dataset.from_tensor_slices(inputs).batch(batch_size).prefetch(tf.data.AUTOTUNE)

test_ds = make_test_ds(Xh, Xl, Xc)

# 4) Carga el modelo guardado (.keras o .h5)
mc_model = keras.models.load_model("mc_dropout_model.keras", compile=False)

# 5) Inferencia MC Dropout lote a lote para capturar incertidumbre
N = 50
all_preds = []

for i in range(N):
    preds_batches = []
    for batch in test_ds:
        # batch es un tuple: (Xh_batch, Xl_batch, cat1_batch,..., cat5_batch)
        preds_batch = mc_model(batch, training=True)             # dropout activo
        preds_batches.append(preds_batch.numpy().flatten())
    all_preds.append(np.concatenate(preds_batches, axis=0))

# Transforma a array (N, n_samples)
all_preds = np.stack(all_preds, axis=0)
mean_preds = all_preds.mean(axis=0)
std_preds  = all_preds.std(axis=0)

# 6) Adjunta resultados al DataFrame y guarda
df_out = df_test.copy()
df_out['pred_mean'] = mean_preds
df_out['pred_std']  = std_preds

df_out.to_csv("test_with_uncertainty.csv", index=False)
print(df_out[['pred_mean','pred_std']].head())
````
