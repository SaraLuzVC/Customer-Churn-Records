```python
df = pd.read_csv("./../data/train.csv")
df.head()
```

```python
import pandas as pd
import numpy as np

# Function to calculate entropy
def entropy(y):
    p = y.value_counts(normalize=True)  # Probability of each class
    return -np.sum(p * np.log2(p + 1e-9))  # Small value to avoid log(0)

# Function to calculate information gain for a feature
def information_gain(df, feature, target="Exited"):
    total_entropy = entropy(df[target])  # Entropy before split
    
    # Entropy after split
    values = df[feature].unique()
    weighted_entropy = sum((len(df[df[feature] == v]) / len(df)) * entropy(df[df[feature] == v][target]) for v in values)
    
    # Information Gain
    return total_entropy - weighted_entropy

# List of features to evaluate
features = ["CreditScore", "Geography", "Gender", "Age", "Tenure", 
            "Balance", "NumOfProducts", "HasCrCard", "IsActiveMember", "EstimatedSalary"]

# Discretize numerical features for faster computation
num_bins = 5  # Reduce unique values to speed up calculations
numerical_features = ["CreditScore", "Age", "Balance", "EstimatedSalary"]

for col in numerical_features:
    df[col] = pd.qcut(df[col], q=num_bins, duplicates="drop").astype(str)

# Compute Information Gain for each feature
ig_results = {feature: information_gain(df, feature) for feature in features}

# Convert results to DataFrame and sort
ig_df = pd.DataFrame(list(ig_results.items()), columns=["Feature", "Information Gain"]).sort_values(by="Information Gain", ascending=False)

# Display results
print(ig_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ig_df = ig_df.sort_values(by="Information Gain", ascending=False)

# Define cold colors (shades of blue)
colors = plt.cm.Blues(np.linspace(0.4, 0.9, len(ig_df)))

# Plot Information Gain
plt.figure(figsize=(10, 6))
plt.barh(ig_df["Feature"], ig_df["Information Gain"], color=colors)
plt.xlabel("Information Gain", fontsize=12)
plt.ylabel("Feature", fontsize=12)
plt.title("Feature Importance based on Information Gain", fontsize=14)
plt.gca().invert_yaxis()  # Highest IG at the top

# Show plot
plt.show()
```
